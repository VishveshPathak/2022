```python
import math
piles = [8011,8387,6007,1235,5595,138,3136,1740,963,9412,802,4475,9695,3713,1742,8559,2237,4356,4012,3449,990,6930,523,931,5937,5902,2817,4088,385,1359,1888,1106,416,670,347,6113,4190,2094,2575,3011,8571,32,6318,9658,708,4061,2481,595,69]
k = 93
status_of_piles = []
list_of_subtractions = []
for i in range(k):
    temp=[]
    for j in range(len(piles)):
        temp.append(0)
    status_of_piles.append(temp)

for i in range(k):
    if i == 0:
        for j in range(len(piles)):
            status_of_piles[i][j] = piles[j]
        max_ = max(piles)
        for j in range(len(status_of_piles[0])):
            if status_of_piles[i][j] == max_:
                status_of_piles[i][j] = status_of_piles[i][j] - math.floor(max_ / 2)
                list_of_subtractions.append(math.floor(max_ / 2))
                break
    else:
        for j in range(len(status_of_piles[i-1])):
            status_of_piles[i][j] = status_of_piles[i-1][j]
        max_ = max(status_of_piles[i-1])
        for j in range(len(status_of_piles[i-1])):
            if status_of_piles[i][j] == max_:
                status_of_piles[i][j] = status_of_piles[i][j] - math.floor(max_ / 2)
                list_of_subtractions.append(math.floor(max_ / 2))
                break
print(sum(piles) - sum(list_of_subtractions))
```
